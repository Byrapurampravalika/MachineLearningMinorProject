```python
# 2.Which country has the most suspected fraud ?
```

```python
	highfraud=df[(df['Order Status']=='SUSPECTED_FRAUD') & (df['Type']=='TRANSFER')]
```

```python
highfraud=highfraud[['Order Region']]
```

```python
highfraud
```

```python
valuecounts=highfraud['Order Region'].value_counts()
```

```python
z1=valuecounts.to_dict()
```

```python
highfraud['fraudcounts'] = highfraud['Order Region'].map(z1)
```

```python
highfraud.sort_values('fraudcounts').tail(1)
```

```python
	import matplotlib.pyplot as plt
	highfraud = df[(df['Order Status'] == 'SUSPECTED_FRAUD') & (df['Type'] == 'TRANSFER')]
	fraud=highfraud['Order Region'].value_counts().plot.pie(figsize=(24,12),startangle=180,autopct='%.1f')
	plt.title("Regions with Highest Fraud ",size=25,color='b')
	plt.ylabel(" ")
	fraud.axis('equal') 
	plt.show()

```
